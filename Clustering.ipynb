## Exercise 1

In this exercise, we will use a bike sharing dataset. The dataset contains coordinates of bike sharing stations in New York given as longitude and latitude values.

The dataset is an example dataset from the book ["Introduction to R for Business Intelligence"](https://jgendron.github.io/com.packtpub.intro.r.bi/), Packt Publishing Ltd., 2016, by Jay Gendron. The dataset is downloadable from Github [here](https://github.com/jgendron/com.packtpub.intro.r.bi/tree/master/Chapter5-ClusteringAnalysis/data), but is also available on Moodle as "Ch5_bike_station_locations.csv".

Answer the following questions:


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler

from sklearn.cluster import KMeans


data=pd.read_csv('Ch5_bike_station_locations.csv')
data.head()
```

```python
x=data[['latitude','longitude']].values
```

1. Apply K-Means clustering to the dataset. Try out different values for K and use the Elbow method to find the best (or a sensible) K.


```python
Kmeans=KMeans(n_clusters=5)
Kmeans.fit(x)
```

```python
Kmeans.labels_
```

```python
Kmeans.cluster_centers_
```

```python
data['5Means_Cluster']=Kmeans.labels_
```

```python
sns.scatterplot(data=data,x='latitude',y='longitude',hue='5Means_Cluster')
plt.scatter(x=Kmeans.cluster_centers_[:,0],y=Kmeans.cluster_centers_[:,1],c='red',s=100)
plt.title("5-Means clustering of the bike station locations")
plt.show()
```

```python
inertia = []
K_range = range(1, 15)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)

plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
best_k = 5
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(x)
```

2. Visualize the K-Means clustering result for the best K.


```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], c=kmeans_labels, cmap='tab10', s=30)
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            c='black', marker='x', s=200)
plt.title("K-Means Clustering (K=5)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

3. Apply Agglomerative clustering to the dataset. Draw the Dendrogram and use it to find the best (or a sensible) number of clusters.


```python

from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(x, method='ward')

plt.figure(figsize=(10,6))
dendrogram(linked)
plt.title("Dendrogram (Agglomerative Clustering)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg_clusters = 5
agg = AgglomerativeClustering(n_clusters=agg_clusters, linkage='ward')
agg_labels = agg.fit_predict(x)
```

4. Generate this number of clusters with Agglomerative clustering and visualize the clustering.


```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], c=agg_labels, cmap='tab10', s=30)
plt.title("Agglomerative Clustering (5 Clusters)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

5. Apply DBSCAN clustering to the dataset. Try out different values for "eps" and "min_samples" and choose the best values. (Hint: For "eps" you probably have to try out values around 0.01.)


```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.01, min_samples=5)
db_labels = dbscan.fit_predict(x)

unique_clusters = len(set(db_labels)) - (1 if -1 in db_labels else 0)
print("Clusters found:", unique_clusters)
```

```python
dbscan = DBSCAN(eps=0.01, min_samples=5)
db_labels = dbscan.fit_predict(x)
```

6. Visualize the final DBSCAN clustering you decided upon in 5.


```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], c=db_labels, cmap='tab10', s=30)
plt.title("DBSCAN Clustering")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

7. Calculate the Silhouette scores for the three different "best clustering" you found with the three different methods. Also use Adjusted Rand Index to compare the best ones to each other.

```python
from sklearn.metrics import silhouette_score
sil_kmeans = silhouette_score(x, kmeans_labels)
sil_agg = silhouette_score(x, agg_labels)

# DBSCAN must exclude noise (-1)
mask = db_labels != -1
sil_db = silhouette_score(x[mask], db_labels[mask])

print("Silhouette Scores:")
print("K-Means:", sil_kmeans)
print("Agglomerative:", sil_agg)
print("DBSCAN:", sil_db)
```

Interpretation:

Closer to 1 = better

Around 0.3–0.5 typical for geographic clustering

```python
from sklearn.metrics import adjusted_rand_score
ari_k_agg = adjusted_rand_score(kmeans_labels, agg_labels)
ari_k_db = adjusted_rand_score(kmeans_labels, db_labels)
ari_agg_db = adjusted_rand_score(agg_labels, db_labels)

print("Adjusted Rand Index:")
print("KMeans vs Agg:", ari_k_agg)
print("KMeans vs DB:", ari_k_db)
print("Agg vs DB:", ari_agg_db)
```

Interpretation:

1 → identical clustering

0 → random agreement

Negative → worse than random

Using the Elbow method, K=5 was selected for K-Means.

The dendrogram suggested 5 clusters for Agglomerative clustering.

DBSCAN with eps=0.01 and min_samples=5 produced meaningful spatial clusters and detected noise points.

Silhouette scores indicate that K-Means and Agglomerative clustering perform similarly and slightly better than DBSCAN.

Adjusted Rand Index shows high similarity between K-Means and Agglomerative results.
